```python
# Time features (NO conversion needed)
df["Time_of_day_of_accident_hrs"] = df["Time_of_Day_of_Accident"].apply(lambda x: x.hour)

def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 1
    elif 12 <= hour < 17:
        return 2
    elif 17 <= hour < 21:
        return 3
    else:
        return 4

df["Time_of_Day"] = df["Time_of_day_of_accident_hrs"].apply(get_time_of_day)
```

```python
import pandas as pd

# ---------------------------
# 1. Clean text safely (preserve NA)
# ---------------------------
df["At_Fault_Party"] = df["At_Fault_Party"].apply(
    lambda x: x.replace("\n", "").strip() if isinstance(x, str) else x
)

df["At_Fault_Party"] = df["At_Fault_Party"].apply(
    lambda x: " ".join(x.split()) if isinstance(x, str) else x
)

# ---------------------------
# 2. NA indicator
# ---------------------------
df["is_unknown"] = df["At_Fault_Party"].isna().astype(int)

# ---------------------------
# 3. Split into list (multi-label)
# ---------------------------
df["At_Fault_List"] = df["At_Fault_Party"].apply(
    lambda x: [i.strip().lower() for i in x.split(",")] if isinstance(x, str) else []
)

# ---------------------------
# 4. Define grouping logic
# ---------------------------
def check_any(lst, keywords):
    return int(any(any(k in item for k in keywords) for item in lst))


df["is_driver"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, [
        "водитель", 
        "имеющий водительские права"
    ])
)

df["is_pedestrian"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["пешеход"])
)

df["is_cyclist"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["велосипедист", "возчик"])
)

df["is_micro_mobility"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, [
        "самоката", 
        "электрического транспортного средства",
        "мопеда"
    ])
)

df["is_no_license"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["не имеющий водительских прав"])
)

df["is_deprived_license"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["лишенный водительских прав"])
)

df["is_minor"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["несовершеннолетний"])
)

df["is_foreign"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["иностранное лицо"])
)

df["escaped_scene"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["скрывшийся с места дтп"])
)

df["is_service_vehicle"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["специальной службы"])
)

df["is_special_worker"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, [
        "коммунальные службы",
        "дорожные службы",
        "подрядчики"
    ])
)

df["is_animal_owner"] = df["At_Fault_List"].apply(
    lambda x: check_any(x, ["хозяин животного"])
)

# ---------------------------
# 5. Drop helper columns
# ---------------------------
df.drop(columns=["At_Fault_Party", "At_Fault_List"], inplace=True)
```

```python
df["Weather_Conditions"] = df["Weather_Conditions"].apply(
    lambda x: x.strip().lower() if isinstance(x, str) else x
)

# NA indicator (VERY IMPORTANT)
df["weather_unknown"] = df["Weather_Conditions"].isna().astype(int)

df["weather_clear"] = (df["Weather_Conditions"] == "ясно").astype(int)
df["weather_cloudy"] = (df["Weather_Conditions"] == "пасмурно").astype(int)
df["weather_rain"] = (df["Weather_Conditions"] == "дождь").astype(int)
df["weather_snow"] = (df["Weather_Conditions"] == "снег").astype(int)
df["weather_fog"] = (df["Weather_Conditions"] == "туман").astype(int)
```

```python
import pandas as pd

# ---------------------------
# 1. Clean text safely
# ---------------------------
df["Road_Surface_Condition"] = df["Road_Surface_Condition"].apply(
    lambda x: x.strip().lower() if isinstance(x, str) else x
)

# ---------------------------
# 2. Missing indicator
# ---------------------------
df["road_surface_unknown"] = df["Road_Surface_Condition"].isna().astype(int)

# ---------------------------
# 3. Split multi-label
# ---------------------------
df["road_surface_list"] = df["Road_Surface_Condition"].apply(
    lambda x: [i.strip() for i in x.split(",")] if isinstance(x, str) else []
)

df["road_ice_snow_risk"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in [
        "заснеженная",
        "обледеневшая",
        "со снежным накатом"
    ]))
)

df["road_wet"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in ["мокрая"]))
)
df["road_dry"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in ["сухая"]))
)
df["road_under_construction"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in ["ремонт дорог"]))
)
df["road_dirty"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in ["загрязненная"]))
)
df["road_fresh_surface"] = df["road_surface_list"].apply(
    lambda x: int(any(i in x for i in ["свежеуложенная"]))
)
```

```python
df.drop(columns=["Road_Surface_Condition", "road_surface_list"], inplace=True)
```

```python
df["Lighting_Conditions"] = df["Lighting_Conditions"].apply(
    lambda x: x.strip().lower() if isinstance(x, str) else x
)

df["External_Lighting_at_Night"] = df["External_Lighting_at_Night"].apply(
    lambda x: x.strip().lower() if isinstance(x, str) else x
)
df["lighting_unknown"] = df["Lighting_Conditions"].isna().astype(int)
df["external_lighting_unknown"] = df["External_Lighting_at_Night"].isna().astype(int)
df["is_day"] = (df["Lighting_Conditions"] == "день").astype(int)
df["is_night"] = (df["Lighting_Conditions"] == "ночь").astype(int)
df["is_twilight"] = (df["Lighting_Conditions"] == "сумерки").astype(int)
df["light_on"] = (df["External_Lighting_at_Night"] == "включено").astype(int)
df["light_off"] = (df["External_Lighting_at_Night"] == "не включено").astype(int)
df["light_absent"] = (df["External_Lighting_at_Night"] == "отсутствует").astype(int)
df["night_without_lighting"] = (
    (df["is_night"] == 1) &
    (df["light_off"] == 1)
).astype(int)
df["twilight_without_lighting"] = (
    (df["is_twilight"] == 1) &
    (df["light_off"] == 1)
).astype(int)
df["lighting_data_missing_risk"] = (
    (df["lighting_unknown"] == 1) |
    (df["external_lighting_unknown"] == 1)
).astype(int)
df.drop(columns=["Lighting_Conditions", "External_Lighting_at_Night"], inplace=True)
```

```python
location_mapping = {
    'на дороге': 1,
    'в населенном пункте': 2,
    'районного значения': 3,
    'областного значения': 4,
    'областного, районного значения': 5,
    'республиканского значения': 6,
    'международного, республиканского значения': 7
}
df['Accident_Location'] = df['Accident_Location'].map(location_mapping)
```

```python
import numpy as np

valid_speeds = np.array([30, 40, 50, 60, 70, 80, 90, 100, 110, 120, 140])

def clean_speed(x):
    if x == 0:
        return 0  # keep unknown
    elif x < 30:
        return 30  # minimum realistic speed
    elif x in valid_speeds:
        return x  # already valid
    else:
        # round to nearest valid speed
        return valid_speeds[np.abs(valid_speeds - x).argmin()]

df['Speed_Limit_clean'] = df['Speed_Limit'].apply(clean_speed)
df[['Speed_Limit', 'Speed_Limit_clean']].drop_duplicates().sort_values('Speed_Limit')
```
